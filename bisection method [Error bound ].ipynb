```python
import numpy as np
from scipy import optimize
f = lambda x: np.cos(x) - x
r = optimize.fsolve(f, -2)
print("R =", r)
result = f(r)
print("result =",result)
```

```python
import numpy as np
def my_bisection(f,a,b,tol):
  if np.sign(f (a)) == np.sign(f (b)):
    raise Exception(
      "The scalars a and b don't bound a root")
  m =(a+b)/2
  if np.abs(f (m)) < tol:
    return m
  elif np.sign(f (a)) == np.sign(f (m)):
    return my_bisection(f,m,b,tol)
  elif np.sign(f (b)) == np.sign(f (m)):
    return my_bisection(f,m,a,tol)

```

```python
f = lambda x: x*x - 2
a = 0
b = 2
tol = 0.1
r = my_bisection(f,a,b,tol)

result = f(r)
print("result =",result)

print("R =", r)
```
